#### Clean the entire data with dropping NAs

```python
rm_na_emp_compensation_data = emp_compensation_data.dropna()
```

#### Display the co-relation between the features

```python
corrmat = rm_na_emp_compensation_data.corr(method='pearson')

sns.set_style("white")

# Mask for the upper triangle
mask = np.zeros_like(corrmat, dtype=bool)
mask[np.triu_indices_from(mask)] = True

# Set figure size
fig, ax = plt.subplots(figsize=(12, 12))

# Define custom colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap
sns.heatmap(corrmat, mask=mask, cmap=cmap, vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": 1}, annot=True,annot_kws={'size': 10})

plt.tight_layout()
fig.show()
```
